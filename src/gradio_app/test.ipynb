```python
import pandas as pd
import json
import re

def parse_ner_json_to_dataframe(json_str: str) -> pd.DataFrame:
    data = json.loads(json_str)

    tokenized_text = data["tokenized_text"]
    ner_spans = data["ner"]

    records = []
    for start, end, label in ner_spans:
        span_tokens = tokenized_text[start:end + 1]
        span_text = " ".join(span_tokens)

        # split compound labels, e.g., "Refugee Finder <> publisher, Global Internal Displacement Database <> publisher"
        label_parts = [lbl.strip() for lbl in label.split(",")]

        for part in label_parts:
            if "<>" in part:
                entity, relation = map(str.strip, part.split("<>"))
            else:
                entity = span_text
                relation = part.strip()
            records.append({
                "start_index": start,
                "end_index": end,
                "span": span_text,
                "entity": entity,
                "relation": relation
            })

    return pd.DataFrame(records)
```
